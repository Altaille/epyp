```python
def groupQ2(df,grps,isotopes):
    s = pd.Series(np.zeros(len(df.index)))
    for iso,weight in isotopes.items():
        try:
            s += weight*df[grps[0]+f'[{iso}]'+grps[1]]
        except KeyError:
            pass
    return s

DM.add_alias(
    regex=r'(.*)\[\{Q2\}\](.*)',
    alias_fct=lambda df,grps: groupQ2(
        df, grps,
        {'HH':1,'HD':1,'HT':1,'DD':1,'DT':1,'TT':1}
    ),
    assigned_alias_grps=['default']
)
DM.add_alias(
    regex=r'(.*)\[\{H2\}\](.*)',
    alias_fct=lambda df,grps: groupQ2(
        df, grps,
        {'HH':1,'HD':0.5,'HT':0.5}
    ),
    assigned_alias_grps=['default']
)
DM.add_alias(
    regex=r'(.*)\[\{D2\}\](.*)',
    alias_fct=lambda df,grps: groupQ2(
        df, grps,
        {'DD':1,'HD':0.5,'DT':0.5}
    ),
    assigned_alias_grps=['default']
)
DM.add_alias(
    regex=r'(.*)\[\{T2\}\](.*)',
    alias_fct=lambda df,grps: groupQ2(
        df, grps,
        {'TT':1,'HT':0.5,'DT':0.5}
    ),
    assigned_alias_grps=['default']
)


def arrAll(df,grps):
    fvname = lambda i: grps[0]+f'[{i}]'+grps[1]
    i = 1
    s = [df[fvname(i)]]
    while True:
        i += 1
        try:
            s.append(df[fvname(i)])
        except KeyError:
            break
    return s

DM.add_alias(
    regex=r'(.*)\[\{\:\}\](.*)',
    alias_fct=arrAll,
    assigned_alias_grps=['default']
)

display(DM.get_df(vnames=[
    'TIME',
    'TEP.f_ISS.z_inF[HH]',
    'TEP.f_ISS.z_inF[HD]',
    'TEP.f_ISS.z_inF[HT]',
    'TEP.f_ISS.z_inF[DD]',
    'TEP.f_ISS.z_inF[DT]',
    'TEP.f_ISS.z_inF[TT]',
], loc=lambda df: df['TIME'] >= 0)['data'])

display(DM.get_df(vnames=[f'TEP.f_ISS.z_inF[{Q2}]' for Q2 in ['{Q2}','{H2}','{D2}','{T2}']], loc=lambda df: df['TIME'] >= 0)['data'])
```
